### 拆分数据

```python
import pandas as pd
data=pd.read_csv("./data/demo.csv")
data.head(5)
```

```python
x_train=data[:500]
x_test=data[500:]
y_train=x_train["Survived"]
y_test=x_test["Survived"]
del x_train["Survived"]
del x_test["Survived"]
```

### pipe构建

```python
from easymlops import NLPPipeline
from easymlops.table.preprocessing import *
from easymlops.table.encoding import *
from easymlops.nlp.preprocessing import *
```

```python
nlp=NLPPipeline()
nlp.pipe(FixInput())\
   .pipe(FillNa())\
   .pipe(TargetEncoding(cols=["Sex","Ticket","Cabin","Embarked"],y=y_train))\
   .pipe(Lower(cols=["Name"]))\
   .pipe(ReplacePunctuation(cols=["Name"],symbols=""))

x_test_new=nlp.fit(x_train).transform(x_test)
x_test_new.head(5)
```

### 性能、一致性测试

```python
nlp.auto_test(x_test[:10])
```

### 持久化测试
